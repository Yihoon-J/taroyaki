```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
url="https://themaintarot.com/"
maintext=requests.get(url).text
soup=BeautifulSoup(maintext,'html.parser')
```

```python
urls=[]
results=soup.find_all('td', 'has-text-align-center')
for r in results:
    try:
        urls.append(r.find('a').get('href'))
    except:
        pass
```

```python
len(urls)
```

```python
meanings=[]
for u in tqdm(urls):
    soup=BeautifulSoup(requests.get(u).text, 'html.parser')
    res=soup.find('div', 'entry-content').text.replace('(adsbygoogle = window.adsbygoogle || []).push({});','')
    meanings.append(res)
```

```python
for m in meanings:
    print(m, "\n")
```
